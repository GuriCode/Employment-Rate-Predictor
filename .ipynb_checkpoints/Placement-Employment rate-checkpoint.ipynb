```python
import pandas as pd
import numpy as np
from sklearn import tree
import numpy as np
from mypipes import *
%matplotlib inline
```

```python
df=pd.read_csv('Placement_Data_Full_Class.csv')
```

```python
df.head()
```

```python
df["status"].value_counts()
```

```python
df.corr()#we can determine the correlation and as it is pretty less therefore each can be contributed as independent variable
```

```python
df=df.drop(["sl_no","salary"],axis=1)
```

```python
df
```

```python
cat_vars=list(df.select_dtypes(include=['object']).columns)
```

```python
cat_vars
```

```python
cat_vars=[a  for a in cat_vars if a not in ['status']]
```

```python
num_vars=list(df.select_dtypes(exclude=['object']).columns)
```

```python
cat_vars
```

```python
p1=pdPipeline([
    ('cat_select',VarSelector(cat_vars)),
    ('missing_trt',DataFrameImputer()),
    ('create_dummies',get_dummies_Pipe(0))
])

p2=pdPipeline([
    ('num_select',VarSelector(num_vars)),
    ('missing_trt',DataFrameImputer())
])

data_pipe=FeatureUnion([
    ('cat_pipe',p1),
    ('num_pipe',p2)
])
```

```python
x_train=pd.DataFrame(data=data_pipe.fit_transform(df),
                     columns=data_pipe.get_feature_names())
```

```python
data_pipe.get_feature_names()
```

```python
y_train=(df['status']=='Placed').astype(int)
```

```python
y_train
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf=DecisionTreeClassifier()
```

```python
random_search=RandomizedSearchCV(clf,
                                 cv=3,
                                 param_distributions=params,
                                 scoring='roc_auc',
                                 n_iter=5,n_jobs=-1,verbose=20
                                    )
```

```python
random_search.fit(x_train,y_train)
```

```python
def report(results, n_top=3):
    for i in range(1, n_top + 1):
        candidates = np.flatnonzero(results['rank_test_score'] == i)
        for candidate in candidates:
            print("Model with rank: {0}".format(i))
            print("Mean validation score: {0:.5f} (std: {1:.5f})".format(
                  results['mean_test_score'][candidate],
                  results['std_test_score'][candidate]))
            print("Parameters: {0}".format(results['params'][candidate]))
            print("")
```

```python
report(random_search.cv_results_,5)
```

```python
dtree=DecisionTreeClassifier(**{'min_samples_split': 10, 'min_samples_leaf': 15, 'max_depth': 50, 'criterion': 'entropy', 'class_weight': None})
```

```python
import os
os.getcwd()
```

```python
dotfile = open("mytree.dot", 'w')

tree.export_graphviz(dtree,out_file=dotfile,
                     feature_names=x_train.columns,
                    class_names=["0","1"],
                     )
dotfile.close()
```

```python
y_pred=dtree.predict(x_train)
```

```python
accuracy=accuracy_score(y_train, y_pred)
print(accuracy)
```
